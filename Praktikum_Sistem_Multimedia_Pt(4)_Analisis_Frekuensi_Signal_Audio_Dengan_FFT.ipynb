##Fast Fourier Transform (FFT): Transformasi dari Domain Waktu ke Domain Frekuensi

**Fast Fourier Transform (FFT)** adalah algoritma yang digunakan untuk menghitung **Discrete Fourier Transform (DFT)** secara efisien. Algoritma ini menjadi fondasi penting dalam pemrosesan sinyal digital, seperti sinyal audio, karena memungkinkan analisis konten frekuensi dari sinyal dalam domain waktu.

### 💡 Apa Itu DFT?

**Discrete Fourier Transform (DFT)** mengubah sinyal diskrit dari **domain waktu** menjadi **domain frekuensi**. Dengan DFT, kita dapat mengetahui **seberapa besar kontribusi tiap frekuensi** terhadap pembentukan sinyal tersebut.

Rumus DFT secara matematis:
\[
X[k] = \sum_{n=0}^{N-1} x[n] \cdot e^{-j \frac{2\pi}{N}kn}
\]

Keterangan:
- \( x[n] \): sinyal dalam domain waktu
- \( X[k] \): hasil transformasi dalam domain frekuensi
- \( N \): jumlah sampel sinyal
- \( k \): indeks frekuensi ke-k

Hasilnya adalah **spektrum frekuensi**, yaitu informasi tentang magnitudo dan fase dari berbagai komponen frekuensi dalam sinyal.


### ⚙️ Bagaimana FFT Menghitung DFT Secara Efisien?

FFT mengoptimalkan proses DFT dari kompleksitas komputasi **\( O(N^2) \)** menjadi **\( O(N \log N) \)**, menjadikannya jauh lebih cepat dan efisien, terutama untuk jumlah data besar. Salah satu metode yang paling umum digunakan adalah **algoritma Cooley–Tukey**, yang bekerja sebagai berikut:

1. **Pisahkan sinyal** menjadi dua bagian:
   - Nilai pada indeks genap: \( x[0], x[2], x[4], \dots \)
   - Nilai pada indeks ganjil: \( x[1], x[3], x[5], \dots \)

2. **Hitung DFT masing-masing** bagian secara rekursif.

3. **Gabungkan hasilnya** menggunakan identitas Euler:
   \[
   X[k] = E[k] + W_N^k \cdot O[k]
   \]
   \[
   X[k + N/2] = E[k] - W_N^k \cdot O[k]
   \]
   di mana \( W_N^k = e^{-j\frac{2\pi k}{N}} \), dan \( E[k] \), \( O[k] \) adalah hasil DFT dari bagian genap dan ganjil.

---

### 🔁 Ilustrasi Transformasi

Misalnya terdapat sinyal waktu:  
\[
x = [1, 2, 3, 4]
\]

Langkah FFT:
- Hitung DFT dari \( [1, 3] \) (genap)
- Hitung DFT dari \( [2, 4] \) (ganjil)
- Gabungkan keduanya menjadi hasil spektrum frekuensi \( X[0], X[1], X[2], X[3] \)

Setiap \( X[k] \) menunjukkan **kontribusi frekuensi ke-k** terhadap sinyal asli, baik dari sisi magnitudo (kuat-lemahnya) maupun fase (pergeseran waktu).

# Model Kompresi Audio dengan FFT

Kompresi audio dengan FFT adalah teknik yang mengubah sinyal audio dari domain waktu ke domain frekuensi, lalu menghilangkan atau mengurangi komponen frekuensi yang tidak terlalu penting, baru kemudian mengubahnya kembali ke domain waktu menggunakan inverse FFT (IFFT). Teknik ini dikenal sebagai lossy compression, karena beberapa informasi dibuang.

````python
import numpy as np
import librosa
import soundfile as sf
import matplotlib.pyplot as plt
from IPython.display import Audio
````

````python
file_path = '/content/classroom.wav'
y, sr = librosa.load(file_path)

display(Audio(y, rate=sr))
````

Jika didengarkan, Audio Raw (classroom.wav), memiliki suara latar (background noise) yang lebih banyak. Suara ambience kelas lebih lengkap, termasuk desis (hiss), gema, atau noise frekuensi tinggi. Kualitas lebih “natural” tapi juga lebih “berisik”.

````python
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr)
plt.title("Gelombang Sinyal Audio")
plt.xlabel("Waktu (detik)")
plt.ylabel("Amplitudo")
plt.show()
````

````python
Y = np.fft.fft(y)
freq = np.fft.fftfreq(len(y), d=1/sr) # d : Jarak antar sampel dalam satuan waktu (detik).
````

**Apa Maksud dari `d = 1 / sr` pada `np.fft.fftfreq()`?**

Pada sintaks ini:

```python
freq = np.fft.fftfreq(len(y), d=1/sr)
```

yang penting adalah parameter `d`, yaitu:

> **Jarak antar sampel dalam satuan waktu (detik).**

---

📏 Penjelasan Konteks

- `sr` adalah **sampling rate**, misalnya 22050 Hz.
  - Artinya: dalam satu detik, sinyal diambil sebanyak **22050 sampel**.
- Maka, **jarak waktu antar sampel** adalah:

        d = 1/sr = 1/22050 = 4.5 x 10^-5 detik

Artinya: **setiap sampel dalam sinyal berjarak sekitar 0.000045 detik satu sama lain.**

---

🔧 Fungsi `np.fft.fftfreq(N, d)`
Fungsi ini menghasilkan array frekuensi dari hasil FFT yang sesuai dengan jumlah sampel `N`, dan jarak antar sampel `d`.

Formula internalnya adalah:
    
    f_k = (k/(N x d)), dengan:

    k = indeks FFT (0, 1, 2, ..., N-1)
    N = panjang sinyal
    d = jarak antar sampel

Jika kita tidak kasih `d`, maka frekuensinya hanya berupa nilai **indeks** (bukan Hz), seperti: `[0, 1, 2, 3, ...]`.

Tapi dengan `d = 1/sr`, hasilnya adalah dalam **satuan Hertz (Hz)**.

---

🔊 Contoh Nyata

Misalnya sinyal `y` berdurasi 1 detik, dan `sr = 22050`.  
Berarti jumlah sampel \( N = 22050 \), dan tiap sampel berjarak \( 1/22050 \) detik.  
Maka:

```python
np.fft.fftfreq(22050, 1/22050)
```

akan menghasilkan array frekuensi:

\[
[0, 1, 2, 3, ..., 11024, -11025, -11024, ..., -1]
\]

yang sesuai dengan sumbu X pada spektrum FFT.

---

Kesimpulan

```python
d = 1 / sr
```

berarti:
> "Saya ingin hasil FFT dinyatakan dalam frekuensi **Hz**, bukan indeks abstrak."

Dan itu penting banget supaya kamu bisa tahu:
- Ada puncak di 440 Hz → mungkin nada A
- Ada noise di atas 15 kHz → bisa difilter


````python
plt.figure(figsize=(12, 4))
plt.plot(freq[:len(freq)//2], np.abs(Y[:len(Y)//2]))
plt.title("FFT Sebelum Kompresi")
plt.xlabel("Frekuensi (Hz)")
plt.ylabel("Magnitudo")
plt.grid(True)
plt.show()
````

````python
threshold = 0.02 * np.max(np.abs(Y))
Y_compressed = np.where(np.abs(Y) > threshold, Y, 0)
````

Langkah kompresi frekuensi dalam kode tersebut bertujuan menyederhanakan sinyal audio dengan mempertahankan hanya frekuensi-frekuensi dominan yang signifikan. Hal ini dilakukan dengan menentukan ambang batas (threshold) berdasarkan 2% dari magnitudo maksimum hasil transformasi Fourier (FFT), yakni `threshold = 0.02 * np.max(np.abs(Y))`, di mana `Y` merupakan hasil FFT dari sinyal audio `y`. Nilai absolut `np.abs(Y)` digunakan untuk memperoleh magnitudo (bukan nilai kompleks), yang mencerminkan seberapa besar kontribusi masing-masing frekuensi terhadap sinyal. Sebagai contoh, jika magnitudo maksimum dalam `Y` adalah 1000, maka threshold-nya adalah 20, artinya hanya frekuensi dengan magnitudo lebih dari 20 yang dipertahankan. Frekuensi-frekuensi yang memiliki magnitudo lebih kecil dari threshold dianggap sebagai noise atau kontribusi minor, lalu diset menjadi nol menggunakan `np.where(np.abs(Y) > threshold, Y, 0)`. Dengan demikian, hanya frekuensi yang memiliki kontribusi signifikan terhadap karakter utama sinyal yang dipertahankan, menghasilkan versi terkompresi dari sinyal yang lebih ringan namun tetap menyerupai bentuk aslinya. Teknik ini bermanfaat dalam pengurangan ukuran data audio, filter noise, dan efisiensi transmisi, dengan tetap mempertahankan informasi akustik yang esensial. Sebagai hasil akhirnya, bentuk gelombang sinyal setelah inverse FFT tetap bisa diputar dan dikenali, meskipun sudah lebih sederhana karena sebagian besar frekuensi kecil telah dihilangkan.

````python
plt.figure(figsize=(12, 4))
plt.plot(freq[:len(freq)//2], np.abs(Y_compressed[:len(Y_compressed)//2]))
plt.title("FFT Sesudah Kompresi")
plt.xlabel("Frekuensi (Hz)")
plt.ylabel("Magnitudo")
plt.grid(True)
plt.show()
````

````python
y_compressed = np.fft.ifft(Y_compressed).real
y_compressed = y_compressed.astype(np.float32)
````

Baris kode `y_compressed = np.fft.ifft(Y_compressed).real` dan `y_compressed = y_compressed.astype(np.float32)` digunakan untuk merekonstruksi sinyal audio dari domain frekuensi setelah dilakukan kompresi pada hasil FFT. Fungsi `np.fft.ifft(Y_compressed)` menghitung **Inverse Fast Fourier Transform (IFFT)**, yang mengubah sinyal kembali ke domain waktu setelah sebagian frekuensi dianggap tidak signifikan dan dihilangkan. Hasil dari IFFT ini bisa mengandung komponen imajiner karena perhitungan numerik, oleh karena itu `.real` digunakan untuk hanya mengambil bagian real-nya, yang merupakan bentuk asli sinyal waktu. Selanjutnya, sinyal ini dikonversi menjadi tipe data `float32` menggunakan `astype(np.float32)` untuk mengurangi penggunaan memori dan memastikan kompatibilitas dengan format audio yang umum, sehingga hasilnya bisa disimpan, diputar, atau diproses lebih lanjut dengan efisien. Dengan demikian, kode ini menyelesaikan proses kompresi dan rekonstruksi sinyal, menjaga kualitas suara yang masih mirip dengan aslinya namun dengan ukuran yang lebih kecil.

````python
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y_compressed, sr=sr)
plt.title("Gelombang Sinyal Sesudah Kompresi")
plt.xlabel("Waktu (detik)")
plt.ylabel("Amplitudo")
````

````python
display(Audio(y_compressed, rate=sr))
````

Audio Terkompresi (classroom compressed.wav), terdengar memiliki noise latar lebih minim. Suara utama (seperti pembicara atau diskusi) terdengar lebih bersih namun agak teredam. Frekuensi tinggi (seperti desis, bunyi kipas, atau gema) banyak yang hilang — ini konsisten dengan grafik FFT.
